Задача 1.
$$f(x)=x_1^2 + 2x_1 x_2 + 3x_2^2 - 2x_1 - 3x_2$$

```python
def makeData ():
    # Строим сетку в интервале от -10 до 10, имеющую 100 отсчетов по обоим координатам
    x = np.linspace(-100, 100, 1000)
    y = np.linspace(-100, 100, 1000)

    # Создаем двумерную матрицу-сетку
    xgrid, ygrid = np.meshgrid(x, y)

    # В узлах рассчитываем значение функции
    #z = numpy.sin (xgrid) * numpy.sin (ygrid) / (xgrid * ygrid)
    z = (xgrid**2) + 2 * xgrid * ygrid + 3 * (ygrid**2) - 2 * xgrid - 3 * ygrid
    return xgrid, ygrid, z


x, y, z = makeData()
fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.plot_surface(x, y, z)
plt.show()
```

Задача 4.
$$f(x) = 90(x_2 - x_1^2)^2 + (2 - x_1)^2$$

```python
def makeData ():
    # Строим сетку в интервале от -10 до 10, имеющую 100 отсчетов по обоим координатам
    x = np.linspace(-100, 100, 1000)
    y = np.linspace(-100, 100, 1000)

    # Создаем двумерную матрицу-сетку
    xgrid, ygrid = np.meshgrid(x, y)

    # В узлах рассчитываем значение функции
    #z = numpy.sin (xgrid) * numpy.sin (ygrid) / (xgrid * ygrid)
    z = 90*(ygrid - xgrid**2)**2 + (2 - xgrid)**2
    return xgrid, ygrid, z


x, y, z = makeData()
fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.plot_surface(x, y, z)
plt.show()
```
